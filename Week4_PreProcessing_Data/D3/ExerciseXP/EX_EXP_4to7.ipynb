```python
# Exercise 6: Create a simple DataFrame and Export
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}

df = pd.DataFrame(data)
df

excel_file_path = '/Users/manuel/Desktop/DI-Bootcamp/Week4_PreProcessing_Data/D3/ExerciseXP/simple_dataframe.xlsx'
json_file_path = '/Users/manuel/Desktop/DI-Bootcamp/Week4_PreProcessing_Data/D3/ExerciseXP/simple_dataframe.json'

#Export excel
df.to_excel(excel_file_path, index=False)

#Export JSON
df.to_json(json_file_path, orient= 'records')

excel_file_path, json_file_path
```
