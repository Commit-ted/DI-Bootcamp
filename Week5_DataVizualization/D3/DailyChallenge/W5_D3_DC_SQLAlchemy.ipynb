```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 2015 dataset
googleplaystore = '/content/googleplaystore.csv'  # Path to the uploaded file
df = pd.read_csv(googleplaystore)

# Display the first few rows of the dataset to inspect its structure
df.head()
```

```python
import pandas as pd
from sqlalchemy import create_engine

# Create an SQLite engine
engine = create_engine('sqlite:///playstore.db')

# Export the dataframe to an SQLite table called 'playstore'
df.to_sql('playstore', con=engine, if_exists='replace', index=False)
```

```python
# Executing the query and fetching results
from sqlalchemy import text # import the text object

# Create a connection
with engine.connect() as conn:
    result = conn.execute(text("SELECT * FROM playstore")) # use the text object to create an executable object from the query string

# Formatting the result as a pandas DataFrame
result_df = pd.DataFrame(result.fetchall(), columns=result.keys())

# Display the first 5 results
result_df.head(5)
```

```python
# Writing and executing the query
query = """
SELECT App, Genres, Size
FROM playstore
WHERE Size > 10 AND Genres NOT LIKE 'Art & Design%'
"""

# Create a connection object from the engine
with engine.connect() as conn:
    result_filtered = conn.execute(text(query)) # execute the query on the connection object

# Formatting the result as a pandas DataFrame
filtered_df = pd.DataFrame(result_filtered.fetchall(), columns=result_filtered.keys())

# Display the first five rows of the filtered DataFrame
filtered_df.head(5)
```

```python
# Display summary statistics for key columns
df_summary = df[['Rating', 'Price']].describe()
df_summary
```

```python
# Ensure all values in the 'Installs' column are strings before cleaning
df['Installs'] = df['Installs'].astype(str)

# Replace unwanted characters (+, ,) and convert to numeric
df['Installs'] = pd.to_numeric(df['Installs'].str.replace('[+,]', '', regex=True), errors='coerce')

# 2. Remove '$' from 'Price' and convert it to numeric
# Replace 'Free' with 0
df['Price'] = df['Price'].astype(str) # Make sure the Price column is of type string
df['Price'] = df['Price'].str.replace('Free', '0')

# Remove '$' and convert to float
df['Price'] = df['Price'].str.replace('$', '', regex=True).astype(float)

# 3. Clean 'Size' by converting it to a uniform numeric value
def convert_size(size_str):
    if 'M' in size_str:
        return float(size_str.replace('M', '')) * 1024  # Convert MB to KB
    elif 'k' in size_str:
        return float(size_str.replace('k', ''))  # Keep KB as it is
    else:
        return None  # For 'Varies with device' and other such values

df['Size'] = df['Size'].apply(convert_size)

# Drop any rows where 'Rating', 'Installs', or 'Price' are missing for analysis
df_cleaned = df.dropna(subset=['Rating', 'Installs', 'Price'])

# EDA 1: Top installed and rated apps
top_installed_apps = df_cleaned[['App', 'Installs', 'Rating', 'Category']].sort_values(by='Installs', ascending=False).head(10)

# EDA 2: Distribution of Free vs Paid Apps
type_distribution = df_cleaned['Type'].value_counts()

# EDA 3: Categories with highest average ratings and installs
category_grouped = df_cleaned.groupby('Category').agg({'Rating': 'mean', 'Installs': 'sum'}).sort_values(by='Rating', ascending=False).head(10)

# Showing the top installed apps, type distribution, and category analysis
top_installed_apps, type_distribution, category_grouped
```
