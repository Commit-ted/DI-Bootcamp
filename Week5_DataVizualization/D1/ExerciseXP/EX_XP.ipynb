```python
#2
import matplotlib.pyplot as plt

# Temperature data for each day of the week
temperature_values = [72, 74, 76, 80, 82, 78, 75]
days_of_week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

# Create a simple line plot
plt.figure(figsize=(8, 5))
plt.plot(days_of_week, temperature_values, marker='o')

# Labeling the axes
plt.xlabel("Day")
plt.ylabel("Temperature (°F)")
plt.title("Temperature Variations Over a Week")

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the file path
file_path = '/Users/manuel/Desktop/DI-Bootcamp/Week5_DataVizualization/D1/ExerciseXP/sales-data-analysis-main/sales_data.csv'

try:
    # Load the dataset
    sales_data = pd.read_csv(file_path)

    # Calculate the total quantity of products sold by the company
    total_quantity_sold = sales_data['quantity'].sum()

    # Identify the category that had the highest revenue and calculate how much revenue it generated
    category_revenue = sales_data.groupby('category')['revenue'].sum()
    highest_revenue_category = category_revenue.idxmax()
    highest_revenue_value = category_revenue.max()

    # Determine the average revenue per sale in the dataset
    average_revenue_per_sale = sales_data['revenue'].mean()

    # Create a bar chart to visualize the total revenue generated in each quarter of the year (Q1, Q2, Q3, Q4)
    sales_data['date'] = pd.to_datetime(sales_data['date'])
    sales_data['Quarter'] = sales_data['date'].dt.to_period('Q')

    quarterly_revenue = sales_data.groupby('Quarter')['revenue'].sum()

    plt.figure(figsize=(10, 6))
    quarterly_revenue.plot(kind='bar')
    plt.title('Total Revenue Generated in Each Quarter')
    plt.xlabel('Quarter')
    plt.ylabel('Total Revenue')
    plt.xticks(rotation=0)
    plt.show()

    # Print the results
    print(f"Total quantity of products sold: {total_quantity_sold}")
    print(f"Category with highest revenue: {highest_revenue_category} (${highest_revenue_value})")
    print(f"Average revenue per sale: ${average_revenue_per_sale:.2f}")

except FileNotFoundError:
    print(f"Error: The file at {file_path} was not found. Please check the path and try again.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Line Plot for y = x^2
x_values = np.arange(-10, 11)
y_values = x_values**2

plt.figure(figsize=(6, 4))
plt.plot(x_values, y_values)
plt.title('Plot of y = x^2')
plt.xlabel('x')
plt.ylabel('y = x^2')
plt.grid(True)
plt.show()

# 2. Bar Chart for Sales Data
products = ['A', 'B', 'C', 'D']
sales = [15, 30, 45, 20]

plt.figure(figsize=(6, 4))
plt.bar(products, sales, color=['blue', 'green', 'red', 'purple'])
plt.title('Sales Data of Products')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()

# 3. Pie Chart for Favorite Fruits
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
percentages = [40, 30, 20, 10]
colors = ['red', 'yellow', 'pink', 'brown']

plt.figure(figsize=(6, 6))
plt.pie(percentages, labels=fruits, colors=colors, autopct='%1.1f%%')
plt.title('Favorite Fruits of a Group of People')
plt.legend(fruits)
plt.show()
```
