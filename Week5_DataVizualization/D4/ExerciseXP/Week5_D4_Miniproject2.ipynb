# Only 2015

```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 2015 dataset
file_2015 = '/content/2015.csv'  # Path to the uploaded file
df_2015 = pd.read_csv(file_2015)

# Display the first few rows of the dataset to inspect its structure
df_2015.head()
```

```python
# Check for missing values
print("Missing Values:")
print(df_2015.isnull().sum())

# Renaming columns for consistency and readability
df_2015.rename(columns={
    'Economy (GDP per Capita)': 'GDP per Capita',
    'Family': 'Social Support',
    'Health (Life Expectancy)': 'Life Expectancy',
    'Freedom': 'Freedom to Make Life Choices',
    'Trust (Government Corruption)': 'Perception of Corruption',
    'Generosity': 'Generosity',
    'Happiness Rank': 'Happiness Rank',
    'Happiness Score': 'Happiness Score'
}, inplace=True)

# Confirming the renaming
df_2015.head()
```

```python
# Summary of key statistics for the 2015 dataset
print(df_2015.describe())

# Checking the distribution of happiness scores
sns.histplot(df_2015['Happiness Score'], bins=20, kde=True)
plt.title('Distribution of Happiness Scores (2015)')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Top 10 happiest countries by Happiness Score
top_10_happiest = df_2015[['Country', 'Happiness Score']].sort_values(by='Happiness Score', ascending=False).head(10)

# Bottom 10 least happy countries by Happiness Score
bottom_10_least_happy = df_2015[['Country', 'Happiness Score']].sort_values(by='Happiness Score', ascending=True).head(10)

# Concatenate top 10 and bottom 10 countries for comparison
top_and_bottom_10 = pd.concat([top_10_happiest, bottom_10_least_happy])

# Bar chart for top and bottom 10 countries
plt.figure(figsize=(10, 6))
sns.barplot(data=top_and_bottom_10, x='Happiness Score', y='Country', palette='coolwarm')
plt.title('Top 10 Happiest and Bottom 10 Least Happy Countries (2015)')
plt.xlabel('Happiness Score')
plt.ylabel('Country')
plt.show()

```

```python
import plotly.graph_objects as go

# Create a figure with an empty layout
fig = go.Figure()

# Add a dropdown menu to choose the region (assuming we have a 'Region' column)
fig.add_trace(go.Bar(
    x=top_10_happiest['Happiness Score'],
    y=top_10_happiest['Country'],
    orientation='h',
    name='Top 10 Happiest'
))

fig.add_trace(go.Bar(
    x=bottom_10_least_happy['Happiness Score'],
    y=bottom_10_least_happy['Country'],
    orientation='h',
    name='Bottom 10 Least Happy'
))

# Add the dropdown for toggling
fig.update_layout(
    updatemenus=[go.layout.Updatemenu(
        buttons=list([
            dict(args=[{'visible': [True, False]}],
                 label="Top 10 Happiest",
                 method="update"),
            dict(args=[{'visible': [False, True]}],
                 label="Bottom 10 Least Happy",
                 method="update")
        ]),
        direction="down"
    )]
)

fig.update_layout(title="Global Happiness Scores (2015)", yaxis={'categoryorder':'total ascending'})
fig.show()
```

```python
# Assuming 'Region' column is available, aggregate by region
region_avg_scores = df_2015.groupby('Region')['Happiness Score'].mean().reset_index()

# Bar chart showing average happiness score by region
plt.figure(figsize=(10, 6))
sns.barplot(data=region_avg_scores, x='Happiness Score', y='Region', palette='deep')
plt.title('Average Happiness Scores by Region (2015)')
plt.xlabel('Average Happiness Score')
plt.ylabel('Region')
plt.show()
```

```python
# Correlation heatmap using Seaborn
corr_matrix = df_2015[['GDP per Capita', 'Social Support', 'Life Expectancy', 'Freedom to Make Life Choices', 'Perception of Corruption', 'Generosity', 'Happiness Score']].corr()

# Plotting the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation between Factors (2015)')
plt.show()
```
