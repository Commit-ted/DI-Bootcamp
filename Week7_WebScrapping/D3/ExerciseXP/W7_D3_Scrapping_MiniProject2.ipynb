```python
# Install Chromedriver
# Get Chrome version
import os
import requests
import re

chrome_version = !google-chrome --version
chrome_version = chrome_version[0].split(' ')[2]
print(f"Chrome version: {chrome_version}")

# Extract the major version number
chrome_major_version = chrome_version.split('.')[0]
print(f"Chrome major version: {chrome_major_version}")

# Get the latest Chromedriver version for the major Chrome version
url = f"https://chromedriver.storage.googleapis.com/LATEST_RELEASE_{chrome_major_version}"
response = requests.get(url)
if response.status_code == 200:
    chromedriver_version = response.text.strip()
    print(f"Chromedriver version: {chromedriver_version}")
else:
    # If the specific major version is not found, get the latest Chromedriver version
    url = "https://chromedriver.storage.googleapis.com/LATEST_RELEASE"
    response = requests.get(url)
    chromedriver_version = response.text.strip()
    print(f"Defaulting to the latest Chromedriver version: {chromedriver_version}")

# Download and install Chromedriver
!wget -N https://chromedriver.storage.googleapis.com/{chromedriver_version}/chromedriver_linux64.zip
!unzip -o chromedriver_linux64.zip -d /usr/local/bin/
!chmod +x /usr/local/bin/chromedriver

# Import modules
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import pandas as pd
from selenium.webdriver.common.by import By

# Set up headless Chrome WebDriver
chrome_options = Options()
chrome_options.add_argument('--headless')  # Run in headless mode
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.binary_location = '/usr/bin/google-chrome'

# Initialize the WebDriver
service = Service('/usr/local/bin/chromedriver')
driver = webdriver.Chrome(service=service, options=chrome_options)

# Navigate to the page
url = 'https://www.scrapethissite.com/pages/frames/'
driver.get(url)

# Initialize data list
data = []

# Get top-level frames
top_frames = driver.find_elements(By.TAG_NAME, 'frame')
print("Number of top-level frames:", len(top_frames))

for i, top_frame in enumerate(top_frames):
    print(f"Processing top-level frame {i+1}: {top_frame.get_attribute('src')}")
    driver.switch_to.frame(top_frame)
    # Now we are inside top_frame
    # Get any nested frames
    nested_frames = driver.find_elements(By.TAG_NAME, 'frame')
    print("Number of nested frames:", len(nested_frames))
    if nested_frames:
        for j, nested_frame in enumerate(nested_frames):
            print(f"Processing nested frame {j+1}: {nested_frame.get_attribute('src')}")
            driver.switch_to.frame(nested_frame)
            # Now we are inside nested frame
            # Get page source
            nested_page_source = driver.page_source
            nested_soup = BeautifulSoup(nested_page_source, 'html.parser')
            # Extract data
            text = nested_soup.get_text()
            data.append({
                'top_frame': top_frame.get_attribute('src'),
                'nested_frame': nested_frame.get_attribute('src'),
                'text': text.strip()
            })
            # Switch back to parent frame (which is top_frame)
            driver.switch_to.parent_frame()
    else:
        # No nested frames
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'html.parser')
        text = soup.get_text()
        data.append({
            'top_frame': top_frame.get_attribute('src'),
            'nested_frame': None,
            'text': text.strip()
        })
    # Switch back to default content
    driver.switch_to.default_content()

# Create DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)

# Save to CSV
df.to_csv('extracted_data.csv', index=False)
print("Data saved to extracted_data.csv")

# Close the driver
driver.quit()
```

```python
# Install necessary packages
!apt-get update
!apt-get install -y firefox-esr
!apt-get install -y wget
!wget -q https://github.com/mozilla/geckodriver/releases/latest/download/geckodriver-v0.33.0-linux64.tar.gz
!tar -xzf geckodriver-v0.33.0-linux64.tar.gz
!chmod +x geckodriver
!mv geckodriver /usr/local/bin/
!pip install selenium
!pip install beautifulsoup4
!pip install pandas

# Import modules
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd

# Set up headless Firefox WebDriver
firefox_options = Options()
firefox_options.add_argument('--headless')
firefox_options.add_argument('--no-sandbox')
firefox_options.add_argument('--disable-dev-shm-usage')

# Initialize the WebDriver
driver = webdriver.Firefox(options=firefox_options)

# Navigate to the page
url = 'https://www.scrapethissite.com/pages/frames/'
driver.get(url)

# Wait for the page to load frames/iframes
wait = WebDriverWait(driver, 10)

# Initialize data list
data = []

# Wait until at least one frame or iframe is present
try:
    wait.until(EC.presence_of_element_located((By.XPATH, "//frame|//iframe")))
except:
    print("No frames or iframes found on the page.")

# Get top-level frames and iframes
top_frames = driver.find_elements(By.XPATH, "//frame|//iframe")
print("Number of top-level frames/iframes:", len(top_frames))

for i, top_frame in enumerate(top_frames):
    frame_src = top_frame.get_attribute('src')
    print(f"Processing top-level frame/iframe {i+1}: {frame_src}")
    driver.switch_to.frame(top_frame)
    # Now we are inside top_frame
    # Wait for any nested frames/iframes to load
    try:
        wait.until(EC.presence_of_element_located((By.XPATH, "//frame|//iframe")))
    except:
        pass
    nested_frames = driver.find_elements(By.XPATH, "//frame|//iframe")
    print("Number of nested frames/iframes:", len(nested_frames))
    if nested_frames:
        for j, nested_frame in enumerate(nested_frames):
            nested_frame_src = nested_frame.get_attribute('src')
            print(f"Processing nested frame/iframe {j+1}: {nested_frame_src}")
            driver.switch_to.frame(nested_frame)
            # Now we are inside nested frame
            # Get page source
            nested_page_source = driver.page_source
            nested_soup = BeautifulSoup(nested_page_source, 'html.parser')
            # Extract data
            text = nested_soup.get_text()
            links = [a['href'] for a in nested_soup.find_all('a', href=True)]
            data.append({
                'top_frame': frame_src,
                'nested_frame': nested_frame_src,
                'text': text.strip(),
                'links': links
            })
            # Switch back to parent frame (which is top_frame)
            driver.switch_to.parent_frame()
    else:
        # No nested frames
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'html.parser')
        text = soup.get_text()
        links = [a['href'] for a in soup.find_all('a', href=True)]
        data.append({
            'top_frame': frame_src,
            'nested_frame': None,
            'text': text.strip(),
            'links': links
        })
    # Switch back to default content
    driver.switch_to.default_content()

# Create DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)

# Save to CSV
df.to_csv('extracted_data.csv', index=False)
print("Data saved to extracted_data.csv")

# Close the driver
driver.quit()
```
